```python
# Initiate Environment
import sqlite3
```

```python
# Connecting to the SQLite database
connection = sqlite3.connect("nghia_dannydiner.db")

# Cursor
cursor = connection.cursor()

# SQL commands to create tables and insert data
sql_commands = [
    """
    CREATE TABLE sales (
        customer_id TEXT,
        order_date TEXT,
        product_id INTEGER
    )
    """,
    """
    INSERT INTO sales
        (customer_id, order_date, product_id)
    VALUES
        ('A', '2021-01-01', 1),
        ('A', '2021-01-01', 2),
        ('A', '2021-01-07', 2),
        ('A', '2021-01-10', 3),
        ('A', '2021-01-11', 3),
        ('A', '2021-01-11', 3),
        ('B', '2021-01-01', 2),
        ('B', '2021-01-02', 2),
        ('B', '2021-01-04', 1),
        ('B', '2021-01-11', 1),
        ('B', '2021-01-16', 3),
        ('B', '2021-02-01', 3),
        ('C', '2021-01-01', 3),
        ('C', '2021-01-01', 3),
        ('C', '2021-01-07', 3)
    """,
    """
    CREATE TABLE menu (
        product_id INTEGER,
        product_name TEXT,
        price INTEGER
    )
    """,
    """
    INSERT INTO menu
        (product_id, product_name, price)
    VALUES
        (1, 'sushi', 10),
        (2, 'curry', 15),
        (3, 'ramen', 12)
    """,
    """
    CREATE TABLE members (
        customer_id TEXT,
        join_date TEXT
    )
    """,
    """
    INSERT INTO members
        (customer_id, join_date)
    VALUES
        ('A', '2021-01-07'),
        ('B', '2021-01-09')
    """
]

# Execute each SQL command
for command in sql_commands:
    cursor.execute(command)

# Commit the changes
connection.commit()

# Close the cursor and the connection
cursor.close()
connection.close()
```
